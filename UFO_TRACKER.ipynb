PROJECT UFO TRACKER

```python
# UFO Tracker by [ANUJ BAGHEL]
import pandas as pd
import matplotlib.pyplot as plt

# My UFO sightings data
ufo_data = [
    {"time": "00:01", "speed": "super fast", "direction": "left", "visible": "yes"},
    {"time": "00:02", "speed": "super fast", "direction": "up", "visible": "no"},
    {"time": "00:03", "speed": "fast", "direction": "right", "visible": "yes"}
]

# Make it a table
ufo_table = pd.DataFrame(ufo_data)
print("UFO Sightings Table:")
print(ufo_table)

# Predict next visibility (basic rule: flip last state)
last_visible = ufo_table["visible"].iloc[-1]
prediction = "no" if last_visible == "yes" else "yes"
print(f"At 00:04, UFO will be: {prediction}")

# Add prediction to table and plot
ufo_table_with_pred = pd.concat([ufo_table, pd.DataFrame({"time": ["00:04"], "speed": ["unknown"], "direction": ["unknown"], "visible": [prediction]})])
plt.plot(ufo_table_with_pred["time"], ufo_table_with_pred["visible"].replace({"yes": 1, "no": 0}), marker="o", color="blue")
plt.title("UFO Visibility Tracker")
plt.xlabel("Time")
plt.ylabel("Visible (1 = Yes, 0 = No)")
plt.grid(True)  # Adds a grid for clarity
plt.show()
```
